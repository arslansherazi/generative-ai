# Auto Gen

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import os
from autogen import AssistantAgent, UserProxyAgent, GroupChat, GroupChatManager
```

```python
openai_api_key = os.environ.get("OPENAI_API_KEY")
if not openai_api_key:
    raise ValueError("Please set the OPENAI_API_KEY environment variable")

llm_config = {
    "config_list": [
        {
            "model": "gpt-4",
            "api_key": openai_api_key,
        }
    ]
}
```

```python
# Agent 1: Idea generator
idea_agent = AssistantAgent(
    name="IdeaAgent",
    system_message="You are a creative brainstorming expert. Generate innovative ideas in response to the user's request.",
    llm_config=llm_config,
)
```

```python
# Agent 2: Critic/Refiner
critic_agent = AssistantAgent(
    name="CriticAgent",
    system_message="You are a critical thinker. Evaluate and refine the ideas proposed by others to make them better.",
    llm_config=llm_config,
)
```

```python
# UserProxyAgent: You (the user)
user_proxy = UserProxyAgent(
    name="User",
    human_input_mode="NEVER",
    llm_config=llm_config,

)
```

```python
group_chat = GroupChat(
    agents=[user_proxy, idea_agent, critic_agent],
    messages=[],
    max_round=10,
)
```

```python
# Create a manager to manage the group conversation
manager = GroupChatManager(
    groupchat=group_chat,
    llm_config=llm_config,
)
```

```python
response = user_proxy.initiate_chat(
    manager,
    message="I want ideas for a mobile app that helps people reduce screen time."
)

print(response)
```
